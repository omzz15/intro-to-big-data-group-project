```python
import pandas as pd
import numpy as np
import os

DATA_DIR = '../../data'
RAW_DATA_DIR = os.path.join(DATA_DIR, 'raw')
CLEANED_DATA_DIR = os.path.join(DATA_DIR, 'cleaned')

STATION_LOCATION_RAW_DATA_DIR = os.path.join(RAW_DATA_DIR, 'station-location')
STATION_LOCATION_CLEANED_DATA_DIR = os.path.join(CLEANED_DATA_DIR, 'station-location')
```

```python
df = pd.read_csv(os.path.join(STATION_LOCATION_RAW_DATA_DIR, "station-locations.csv"), index_col=0)
df
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Save cleaned data
os.makedirs(STATION_LOCATION_CLEANED_DATA_DIR, exist_ok=True)

df.to_csv(os.path.join(STATION_LOCATION_CLEANED_DATA_DIR, "station-locations.csv"), index=False)
```
