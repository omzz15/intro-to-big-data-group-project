Data from https://gadm.org/download_country.html (Level 2 GeoJSON for South Korea)

```python
import geopandas as gpd
import os

DATA_DIR = '../../data'
RAW_DATA_DIR = os.path.join(DATA_DIR, 'raw')
CLEANED_DATA_DIR = os.path.join(DATA_DIR, 'cleaned')

DISTRICT_LOCATION_RAW_DATA_DIR = os.path.join(RAW_DATA_DIR, 'district-location')
DISTRICT_LOCATION_CLEANED_DATA_DIR = os.path.join(CLEANED_DATA_DIR, 'district-location')
```

```python
gdf = gpd.read_file(os.path.join(DISTRICT_LOCATION_RAW_DATA_DIR, 'gadm41_KOR_2.json'))
gdf
```

```python
# Check for missing values
gdf.isnull().sum()
```

```python
# Looks good! Just save it as CSV
os.makedirs(DISTRICT_LOCATION_CLEANED_DATA_DIR, exist_ok=True)

gdf.to_file(os.path.join(DISTRICT_LOCATION_CLEANED_DATA_DIR, "district-location.geojson"), driver='GeoJSON')
print("District location data cleaned and saved!")
```
