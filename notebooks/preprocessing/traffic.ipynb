```python
import pandas as pd
import numpy as np
import os

DATA_DIR = '../../data'
RAW_DATA_DIR = os.path.join(DATA_DIR, 'raw')
CLEANED_DATA_DIR = os.path.join(DATA_DIR, 'cleaned')

TRAFFIC_RAW_DATA_DIR = os.path.join(RAW_DATA_DIR, 'traffic')
TRAFFIC_CLEANED_DATA_DIR = os.path.join(CLEANED_DATA_DIR, 'traffic')
```

```python
traffic_df = pd.read_csv(os.path.join(TRAFFIC_RAW_DATA_DIR, '영업용자동차_업종별_수송_20260115150836.csv'), encoding='euc-kr')
display(traffic_df.columns)
display(traffic_df)
```

```python
cleaned_traffic_df = traffic_df.copy()

# Rename columns to english
cleaned_traffic_df.columns = ['Month', 'Year', 'Industry_1', 'Industry_2', 'Num_Registered_Vehicles', 'Num_Transported_Persons', 'Transport_volume_Tons']

# Convert Industry to english
industry_1_korean = ['여객', '화물']
industry_1_english = ['Passenger', 'Cargo']
cleaned_traffic_df.replace({'Industry_1': dict(zip(industry_1_korean, industry_1_english))}, inplace=True)

industry_2_korean = ['소계', '시내버스', '시외버스', '택시', '전세', '일반화물', '개인(개별)화물', '개인(용달)화물', '개인화물']
industry_2_english = ['Subtotal', 'City Bus', 'Intercity Bus', 'Taxi', 'Charter', 'General Cargo', 'Personal (Individual) Cargo', 'Personal (Delivery) Cargo',
'Personal Cargo']
cleaned_traffic_df.replace({'Industry_2': dict(zip(industry_2_korean, industry_2_english))}, inplace=True)

# Replace - with NaN
cleaned_traffic_df.replace({'Num_Registered_Vehicles': {'-': np.nan},
                            'Num_Transported_Persons': {'-': np.nan},
                            'Transport_volume_Tons': {'-': np.nan}}, inplace=True)

cleaned_traffic_df
```
